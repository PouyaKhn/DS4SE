# 1.E
MAE Mean Absolute Error: میانگین قدرمطلق تفاوت مقدار واقعی و مقدار پیش بینی شده را میدهد, در واقع بزرگی خطا را نشان میدهد (بدون جهت)
 

MSE Mean Squared Error: میانگین مجذور تفاوت مقدار واقعی و مقدار پیش بینی شده

RMSE Root Mean Squared Error: از مقدار بالا ریشه دوم بگیریم

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_error
import numpy as np

print('MAE', mean_absolute_error(y_test, linear_predicted))
print('MSE', mean_squared_error(y_test, linear_predicted))
print('RMSE', np.sqrt(mean_squared_error(y_test, linear_predicted)))
```

# 2.A.b

# 2.B

```python
df.replace(['x', 'o', 'b'], [1, -1, 0], inplace = True)
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
x_train_connect, x_test_connect, y_train_connect, y_test_connect = train_test_split(x, y,random_state=0 ,test_size=0.2, stratify = y )
```

# 2.C

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier

accuracies = []
names = []
def estimate_accuray(model, model_name):
    skf = StratifiedKFold(n_splits = 10)
    score_list = cross_val_score(model, x_train_connect, y_train_connect, scoring='accuracy', n_jobs=-1, cv = skf)
    print(model_name, score_list.mean() * 100)
    accuracies.append(score_list)
    names.append(model_name)
    

models = [{'model_name': 'SVC', 'model_instance': SVC()},
          {'model_name': 'L SVC', 'model_instance': LinearSVC()},
          {'model_name': 'DTC', 'model_instance': DecisionTreeClassifier()},
          {'model_name': 'G Nb', 'model_instance': GaussianNB()},
          {'model_name': 'B Nb', 'model_instance': BernoulliNB()},
          {'model_name': 'KNC', 'model_instance': KNeighborsClassifier()}]


for model in models:
    estimate_accuray(model['model_instance'], model['model_name'])
```

# 2.D

```python
plt.boxplot(accuracies, labels = names)
plt.show()
```

# 2.E

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, plot_confusion_matrix
svc = SVC()
svc.fit(x_train_connect, y_train_connect)
plot_confusion_matrix(svc, x_test_connect, y_test_connect)
print("this confusion matrix is predicting target values by itself,we don't need to predict by ourselves ,it's imported from sklearn.metrics.plot_confusion_matrix")
plt.show()
```

# 3.A

```python
df_fish = pd.read_csv("Fish.csv")
x_fish = df_fish.iloc[:, 1:]
y_fish = df_fish.iloc[:, 0]
x_train_fish, x_test_fish, y_train_fish, y_test_fish = train_test_split(x_fish, y_fish,random_state=0 ,test_size=0.2)
```

# 3.B 3.C 3.D 3.E

# در این سوال به صورت داینامیک با استفاده از یک آرایه برای 3 کرنل متفاوت مطلوبات سوال به دست آمده و نمایش داده شده است

```python
kernels = ['linear', 'rbf', 'poly']
for kernel in kernels:
    fish_svc = SVC(kernel = kernel)
    fish_svc.fit(x_train_fish, y_train_fish)
    pred_fish = fish_svc.predict(x_test_fish)
    print("kernel =", kernel)
    print("classification_report\n", classification_report(y_test_fish, pred_fish, zero_division=0))
    print("accuracy_score\n", accuracy_score(y_test_fish, pred_fish))
    print("confusion matrix\n", confusion_matrix(y_test_fish, pred_fish))
    print("\n\n\n\n\n")
```
